```python
# Create and launch hello world example
import gradio as gr


def greet(name):
    return "Hello " + name


demo = gr.Interface(fn=greet, inputs="text", outputs="text")

demo.launch()
```

```python
# Create and launch hello world example, with explicitly defined inputs
import gradio as gr


def greet(name):
    return "Hello " + name


# We instantiate the Textbox class
textbox = gr.Textbox(label="Type your name here:", placeholder="John Doe", lines=2)

gr.Interface(fn=greet, inputs=textbox, outputs="text").launch()
```

```python
# Instantiate model and create predict function
from transformers import pipeline

model = pipeline("text-generation")


def predict(prompt):
    completion = model(prompt)[0]["generated_text"]
    return completion
```

```python
# Launch application using model above
import gradio as gr

gr.Interface(fn=predict, inputs="text", outputs="text").launch()
```
